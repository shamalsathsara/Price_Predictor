```python
# Strip the 'GB' text before converting
data['Ram'] = data['Ram'].astype(str).str.replace('GB', '', regex=False).astype('int32')

# For Weight, strip 'kg' before converting
data['Weight'] = data['Weight'].astype(str).str.replace('kg', '', regex=False).astype('float32')
```

```python
data.head(2)
```

```python
data.info()
```

```python
data.corr(numeric_only=True)['Price_euros']
```

```python
def add_company(inpt):
    if inpt == 'Samsung' or inpt == 'Razer'or inpt == 'Mediacom'or inpt== 'Microsoft'or inpt=='Xiaomi'or inpt=='Vero'or inpt=='Chuwi'or inpt=='Google ' or inpt=='Fujitsu'or inpt=='LG'or inpt=='Huawei':
        return 'Other'
    else:    
        return inpt

data['Company'] =  data['Company'].apply(add_company)
    
```

```python
data['Company'].value_counts()
```

```python
data['Touchscreen'] = data['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0)
data['Ips']= data['ScreenResolution'].apply(lambda x:1 if 'IPS' in x else 0)
```

```python
data.head(2)
```

```python
data.head(2)
```

```python
data['Cpu'].value_counts()
```

```python
data['Cpu_name']= data['Cpu'].apply(lambda x:" ".join(x.split()[0:3]))
```

```python
data['Cpu_name'].value_counts()
```

```python
def set_processor(name):
    if name == 'Intel Core i7' or name == 'Intel Core i5' or name == 'Intel core i3':
        return name
    else:
        if name.split()[0] == 'AMD':
            return 'AMD'
        else:
            return 'Other'
            
data['Cpu_name'] = data['Cpu_name'].apply(set_processor)            
```

```python
data['Cpu_name'].value_counts()
```

```python
data['Gpu'].value_counts()
```

```python
data['Gpu_name']= data['Gpu'].apply(lambda x:" ".join(x.split()[0:1]))
```

```python
data['Gpu_name'].value_counts()
```

```python
data.shape
```

```python
data = data[data['Gpu_name'] != 'ARM']
```

```python
data.shape
```

```python
data.head(2)
```

```python
data['OpSys'].value_counts()
```

```python
def set_os(inpt):
    if inpt == 'Windows 10' or inpt == 'Windows 7' or inpt == 'Windows 10 S':
        return 'Windows'
    elif inpt == 'macOS' or inpt =='Mac OS X':
        return 'Mac'
    elif inpt == 'Linux':
        return inpt
    else:
        return 'Other'

            
data['OpSys'] = data['OpSys'].apply(set_os) 
```

```python
data.head()
```

```python
data= data.drop(columns=['laptop_ID', 'Inches', 'Product', 'Cpu', 'Gpu'])
```

```python
data.head(2)
```

```python
data = pd.get_dummies(data)
```

```python
data.head(2)
```

```python
print(data.dtypes)
```

```python
# After creating dummies
data = pd.get_dummies(data, drop_first=True)

# Convert all bool columns to int
data = data.astype(int)
```

```python
data.head(2)
```

```python
X = data.drop('Price_euros', axis=1)
y = data['Price_euros']
```

```python
!pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size=0.25)
```

```python
X_train.shape, X_test.shape
```

```python
def model_acc(model):
    model.fit(X_train,y_train)
    acc = model.score(X_test,y_test)
    print(str(model)+ ' --> '+str(acc))
```

# LinearRegression Accurace Test

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model_acc(lr)
```

# LassoRegression Accuracy Test

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
model_acc(lasso)
```

# DecesionTressRegression Accuracy Test

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
model_acc(dt)
```

# RandomForest Accuracy Test

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
model_acc(rf)
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {'n_estimators':[10,50,100],
              'criterion':['squared_error','absolute_error','poisson']}

grid_obj = GridSearchCV(estimator=rf, param_grid=parameters)
grid_fit = grid_obj.fit(X_train, y_train)
best_model = grid_fit.best_estimator_
best_model
```

```python
best_model.score(X_test, y_test)
```

```python
import pickle
with open('predictor.pickle','wb') as file:
    pickle.dump(best_model, file)
```
